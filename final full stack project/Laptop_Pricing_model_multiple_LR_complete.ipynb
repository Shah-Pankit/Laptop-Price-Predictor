```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error ,  mean_squared_error
from sklearn.preprocessing import OneHotEncoder
import joblib
from math import sqrt
```

```python
data=pd.read_csv("laptop_pricing_dataset.csv")
data
```

```python
columns=['Manufacturer','GPU','CPU_core','RAM_GB','Storage_GB_SSD','Price']
columns
```

```python
new_data=data[columns]
new_data
```

```python
print(new_data.columns)
```

```python
encoder= OneHotEncoder()
```

```python
manufacturer_2d = new_data['Manufacturer'].values.reshape(-1,1)
```

```python
new_manufacturer=encoder.fit_transform(manufacturer_2d)
```

```python
array_encoded_manufacturer = new_manufacturer.toarray()
```

```python
new_data.drop(columns=['Manufacturer'], inplace=True)  # Drop the original column
new_data[encoder.get_feature_names_out(['Manufacturer'])] = array_encoded_manufacturer  # Add the encoded columns

# Optionally, you can convert the encoded columns to integers if needed
new_data[encoder.get_feature_names_out(['Manufacturer'])] = new_data[encoder.get_feature_names_out(['Manufacturer'])].astype(int)

# Display the updated DataFrame
print(new_data)
print(new_data.columns)
```

```python
print(new_data.columns)
```

```python
x=new_data[['GPU', 'CPU_core', 'RAM_GB', 'Storage_GB_SSD', 'Manufacturer_Acer',
       'Manufacturer_Asus', 'Manufacturer_Dell', 'Manufacturer_HP',
       'Manufacturer_Huawei', 'Manufacturer_Lenovo', 'Manufacturer_MSI',
       'Manufacturer_Razer', 'Manufacturer_Samsung', 'Manufacturer_Toshiba',
       'Manufacturer_Xiaomi']]
       
```

```python
y=new_data['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=3/4,random_state=0)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
mse=mean_absolute_error(y_pred,y_test)
mse
```
